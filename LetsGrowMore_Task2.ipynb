```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

### READING DATA 

```python
data = pd.read_csv('globalterrorismdb_0718dist.csv',low_memory=False)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.columns
```

```python
# Renamming File Names

data.rename(columns={'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'}, inplace=True)
```

```python
# Removing Unnecessary Data

data=data[['Year','Month','Extended','Day','Country','State','Region','City','Latitude','Longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
data.describe()
```

```python
#Updated Data
data.head()
```

```python
#Terrorism affected data

data['Region'].value_counts()
```

## DATA VISUALIZATION



## Terrorist Activities by Region each year



```python
pd.crosstab(data.Year, data.Region).plot(kind='line',figsize=(22,10))
plt.title('Terrorist Activities by Region each year')
plt.ylabel('Count Number of Attacks')
plt.grid()
plt.xticks([1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2017])
```

## Plotting Barplot


```python
plt.figure(figsize = (20,5))
sns.barplot(data['Region'].value_counts()[:15].index,data['Region'].value_counts()[:15].values,palette='tab10')
plt.title('Most Affected Region')
plt.xlabel('Regions')
plt.ylabel('Count')
plt.xticks(rotation = 90)
```

#### Conclusion 1:- Most Affected Region->Middle East & North Africa

#### Conclusion 2: Least Affected Region->Australasia & Oceania


### Top 10 Affected Countries: 

```python
plt.figure(figsize = (20,5))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values)
plt.title('Top 10 Affected Countries')
plt.xlabel('Countries')
plt.ylabel('Counts')
plt.xticks(rotation = 90)
```

#### Conclusion 1: Most Affected Country-> Iraq

#### Conclusion 2: Least Affected Country -> Turkey



```python
# Overall States
data['State'].value_counts().head(10)
```

```python
plt.figure(figsize = (20,6))
sns.barplot(data['State'].value_counts()[:10].index,data['State'].value_counts()[:10].values)
plt.title('Top 10 Affected States (Overall)')
plt.xlabel('States')
plt.ylabel('Counts')
plt.xticks(rotation = 90)
```

#### Conclusion: Most Affected state = Baghdad

#  Number of Terrorist Activities each Year

```python
data['Year'].value_counts()
```

```python
#Plotiing Graphs
plt.figure(figsize = (20,10))
sns.countplot('Year',data=data,palette='tab10')
plt.xticks(rotation = 90)
plt.title('Number of Terrorist Activities each Year')
plt.xlabel('Year')
plt.ylabel('Counts')
```

#### Conclusion: Highest number of attacks in 2014
#### Conclusion: Lowest number of attacks in 1971





## Attack Methods

```python
# Methods of Attack
plt.figure(figsize = (20,6))
sns.countplot(data['AttackType'],order = data['AttackType'].value_counts().index,palette='tab10')
plt.title('Methods of Attack')
plt.xlabel('Methods')
plt.ylabel('Counts')
plt.xticks(rotation = 90)
```

#### Conclusion 1: Most of the attack types = Bombing/Explosion
#### Conclusion 2: Least of the attack types= Hijacking





## Types of Target

```python
plt.figure(figsize = (20,6))
sns.countplot(x="Target_type", data=data, order=data["Target_type"].value_counts().index)
plt.title('Types of Targets')
plt.xticks(rotation = 90)
```

#### Conclusion: Most Target type = Private Citizens & Property



## Most Frequent Groups Invloved in Terrorist Activity

```python
data['Group'].value_counts().head(10)
```

```python
plt.subplots(figsize = (20,10))
sns.barplot(y=data['Group'].value_counts()[:10].index, x=data['Group'].value_counts()[:10].values, palette='tab10')
plt.title('Most Active Terrorist Organizations')
```
